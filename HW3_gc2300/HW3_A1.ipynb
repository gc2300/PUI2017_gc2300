```python
# 100 samples of different sized N (10<N<2000) from a Normal, a Poisson, a Binomial,
# a Chi-Squared distribution and a uniform distribution, all with the same population mean
from random import radint
```

```python
#Binomial distribution
np.random.seed(0)
d1=[]
mean1=[]
size1=[]
#mean is 20
for i in range(100):
    N = randint(10,2000)
    d1.append(np.random.binomial(100,0.2,N))
    mean1.append(d1[i].mean())
    size1.append(N)
plot1 = plt.figure(figsize=(10,6)).add_subplot(111)
plot1.scatter(size1, mean1)
plot1.set_ylabel("Sample mean", fontsize = 20)
plot1.set_xlabel("Sample size", fontsize = 20)
plot1.set_title("Binomial n=%d, p=%.1f"%(100, 0.2), color="IndianRed", fontsize=18)
```

```python
#Poisson distribution
np.random.seed(0)
d2=[]
mean2=[]
size2=[]
for i in range(100):
    N = randint(10,2000)
    d2.append(np.random.poisson(20,N))
    mean2.append(d2[i].mean())
    size2.append(N)
plot2 = plt.figure(figsize=(10,6)).add_subplot(111)
plot2.scatter(size2, mean2)
plot2.set_ylabel("Sample mean", fontsize = 20)
plot2.set_xlabel("Sample size", fontsize = 20)
plot2.set_title("Poisson lam=%d"%(20), color="IndianRed", fontsize=18)
```

```python
#Normal distribution, standard deviation is 0.5
np.random.seed(0)
d3=[]
mean3=[]
size3=[]
for i in range(100):
    N = randint(10,2000)
    d3.append(np.random.normal(20,0.5,N))
    mean3.append(d3[i].mean())
    size3.append(N)
plot3 = plt.figure(figsize=(10,6)).add_subplot(111)
plot3.scatter(size3, mean3)
plot3.set_ylabel("Sample mean", fontsize = 20)
plot3.set_xlabel("Sample size", fontsize = 20)
plot3.set_title("Normal mean=%d, standard deviation = %f"%(20, 0.5), color="IndianRed", fontsize=18)
```

```python
# Chi-square distribution
np.random.seed(0)
d4=[]
mean4=[]
size4=[]
for i in range(100):
    N = randint(10,2000)
    d4.append(np.random.chisquare(20,N))
    mean4.append(d4[i].mean())
    size4.append(N)
plot4 = plt.figure(figsize=(10,6)).add_subplot(111)
plot4.scatter(size4, mean4)
plot4.set_ylabel("Sample mean", fontsize = 20)
plot4.set_xlabel("Sample size", fontsize = 20)
plot4.set_title("Chi square df=%d"%(20), color="IndianRed", fontsize=18)
```

```python
# uniform distribution
np.random.seed(0)
d5=[]
mean5=[]
size5=[]
for i in range(100):
    N = randint(10,2000)
    d5.append(np.random.uniform(0,40,N))
    mean5.append(d5[i].mean())
    size5.append(N)
plot5 = plt.figure(figsize=(10,6)).add_subplot(111)
plot5.scatter(size5, mean5)
plot5.set_ylabel("Sample mean", fontsize = 20)
plot5.set_xlabel("Sample size", fontsize = 20)
plot5.set_title("Uniform low=%d, high=%d"%(0, 40), color="IndianRed", fontsize=18)
```

```python
#The larger the sample size, the sample mean is closer to population mean.
```

```python
plot6 = plt.figure(figsize=(10,6)).add_subplot(111)
plot6.hist(np.asarray([mean1, mean2, mean3, mean4, mean5]))
plot6.set_xlabel("Sample mean", fontsize = 20)
plot6.set_ylabel("Frequency", fontsize = 20)
plot6.set_title("The distributions of all sample means", color="IndianRed", fontsize=18)
```
