```python
ad = pd.read_csv(os.getenv("PUIDATA")+'/crime.csv')
```

```python
ad = np.transpose(ad)
```

```python
ad
```

```python
crime = ad.iloc[1:19,2:11]
```

```python
crime = crime.iloc[1:18,:]
```

```python
crime
```

```python
del crime.index.name
```

```python
crime
```
